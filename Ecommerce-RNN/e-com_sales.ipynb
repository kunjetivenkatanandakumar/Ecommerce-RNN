```python
import pandas as pd

df = pd.read_csv("ecom_sales_data.csv")

print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

sales = scaler.fit_transform(df[['Sales']])
```

```python
import numpy as np

X = []
y = []

time_step = 3

for i in range(len(sales)-time_step):
    X.append(sales[i:i+time_step])
    y.append(sales[i+time_step])

X = np.array(X)
y = np.array(y)

print(X.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

model = Sequential()

model.add(
    SimpleRNN(
        units=32,
        activation='tanh',
        input_shape=(3,1)
    )
)

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse'
)

model.summary()
```

```python
history = model.fit(
    X,
    y,
    epochs=50,
    batch_size=2
)
```

```python
model.save("sales_rnn_model.keras")

print("Model Saved")
```
